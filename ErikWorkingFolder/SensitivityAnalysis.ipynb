```python
from ema_workbench import (MultiprocessingEvaluator, ema_logging,
                           perform_experiments, SequentialEvaluator)
ema_logging.log_to_stderr(ema_logging.INFO)
 
n_scenarios = 100
n_policies = 10
    
with MultiprocessingEvaluator(model) as evaluator:
    experiments, outcomes = evaluator.perform_experiments(scenarios=n_scenarios, policies=n_policies)
```

```python
from ema_workbench.analysis import pairs_plotting
fig, axes = pairs_plotting.pairs_scatter(experiments=experiments, outcomes=outcomes, group_by='policy',
                                         legend=False)
fig.set_size_inches(16,12)
plt.show()
```

```python
from ema_workbench.analysis import feature_scoring
cleaned_experiments = experiments.drop([u.name for u in model.uncertainties], axis=1)
scores = feature_scoring.get_feature_scores_all(cleaned_experiments, outcomes)
plt.figure(figsize=(16, 12))
sns.set(font_scale=1.5)
sns.heatmap(scores, cmap='viridis', annot=True)
plt.show()
```
