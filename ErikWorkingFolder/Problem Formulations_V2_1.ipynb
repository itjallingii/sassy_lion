```python
def generatestacking(outcome):
    data = outcomes[outcome]

    y = data > np.percentile(data, 95)

    dimensional_stacking.create_pivot_plot(experiments_uncertainties, y,nbins=3, nr_levels=3)
    #plt.savefig('Worst-TotalInvestmentCosts-15000s-1p.png')
    print("This is the Dimensional Stacking visualization for the outcome", outcome)
    plt.show()
```

```python
outcomekeys = outcomes.keys()
for outcome in outcomekeys:
    generatestacking(outcome)
```
